```python
import pandas as pd
from pandas import read_csv
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
# read the data
data = pd.read_csv("/Users/peter/Downloads/Microsoft-Malware-Challenge-master/Dataset/train/LargeTrain.csv")
```

```python
# trans data to np array
dataNpArray = data.values
```

```python
# columns name of data
dataCol = data.columns
```

```python
# train data(X) and label(Y)
X = dataNpArray[:,0:1804]
Y = dataNpArray[:,1804]
```

```python
# random forest
rf = RandomForestRegressor()
rf.fit(X,Y)
```

```python
# feature importance
fi = rf.feature_importances_
```

```python
# class the important feature and not important feature
# more significant feature
MSF = []
MSF_Importance = []
# less significant feature
LSF = []
LSF_Importance = []

cnt = 0
for i in fi:
    # weight > 0.001 => important
    if i >= 0.001:
        MSF.append(dataCol[cnt])
        MSF_Importance.append(i)
        cnt = cnt + 1
    # weight < 0.001 => not important
    else:
        LSF.append(dataCol[cnt])
        LSF_Importance.append(i)
        cnt = cnt + 1
```
